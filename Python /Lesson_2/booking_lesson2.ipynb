```python
# Создайте колонку total_kids, объединив столбцы children и babies
```

```python
bookings['total_kids'] = bookings.children + bookings.adults
```

```python
bookings.groupby('hotel').total_kids.mean().round(2)
```

```python
def has_kids(x):
    if x > 0:
        return True
    else:
        return False
```

```python
bookings['has_kids'] = bookings.total_kids.apply(has_kids)
```

```python
# Посчитаем метрику Churn Rate (отток, коэффициент оттока)
```

```python
bookings[['has_kids', 'is_canceled']].value_counts() / bookings[['has_kids']].value_counts()*100
```
