```python
obj, integr, flt = 0, 0, 0
for i in df_taxi.dtypes:
    if i == 'object':
        obj += 1
    elif i == 'int64':
        integr += 1
    elif i == 'float64':
        flt += 1

print(f'{obj = }, {integr = }, {flt = }')
```

```python
# 6. Выяснить, из какого района было совершено наибольшее количество поездок за весь период.
# Для начала, посчитайте общее количество поездок (pickups), без группировки.
```

```python
df_taxi.value_counts('pickups')
```

```python
df_taxi.pickups.sum()
```

```python
df_taxi.groupby('borough').pickups.sum().sort_values(ascending=False)
```

```python
min_pickups = df_taxi.groupby('borough').pickups.sum().idxmin()
```

```python
min_pickups
```

```python
max_pickups = df_taxi.groupby('borough').pickups.sum().idxmax()
```

```python
max_pickups
```

```python
# Сравните среднее число поездок, и выберите районы, из которых по праздникам 
# в среднем поступает больше заказов, чем в обычные дни. 
```

```python
df_taxi.groupby(['borough', 'hday']).pickups.mean()
# .mean() - среднее значение
```

```python
df_taxi.groupby(['borough', 'hday']).pickups.mean().to_frame()
# .to_frame() - конвертирование в dataframe
```

```python
pickups_by_mon_bor = df_taxi.groupby(['borough', 'pickup_month'], as_index=False) \
                         .agg({'pickups': 'sum'}) \
                         .sort_values('pickups', ascending=False)
```

```python
pickups_by_mon_bor
```
